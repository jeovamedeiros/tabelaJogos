```python
import pandas as pd

tabela = pd.read_csv("planilhaJogos.csv" ,delimiter=';')

tabela= tabela.dropna()

print(tabela.to_string())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
cores = ['gray', 'green', 'black', 'magenta', 'brown']
for i, row in tabela.iterrows():
    plt.bar(row['JOGOS'],row['VALOR GASTO DENTRO DO JOGO'], color=cores[i % len(cores)])
plt.xlabel('JOGOS')
plt.ylabel('VALOR GASTO DENTRO DO JOGO')
plt.title('Tabela de Jogos')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
plt.savefig('graficoJogos.png')
```
